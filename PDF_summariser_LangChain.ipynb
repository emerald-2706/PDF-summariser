```python
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_groq import ChatGroq

# 1. Load the PDF
loader = PyPDFLoader("/content/Instructions (1).pdf") # Upload a PDF to Colab first!
data = loader.load()

# 2. Split into chunks (Crucial for GenAI)
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
chunks = text_splitter.split_documents(data)

# 3. Create Embeddings (Turning text into numbers)
# We use a free model from HuggingFace so you don't need a second API key
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# 4. Store in a local Vector Database
vectorstore = Chroma.from_documents(documents=chunks, embedding=embeddings)

print(f"Successfully split into {len(chunks)} chunks and stored in Chroma.")
```

```python
from langchain_classic.chains import RetrievalQA
from langchain_groq import ChatGroq
import getpass
import os

# 1. Initialize the LLM (The Professor)
# Replace 'YOUR_API_KEY' with your actual Groq key
# This will pop up a secure text box so you don't have to write the key in the cell
os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API Key: ")
llm = ChatGroq(
    model_name="llama-3.1-8b-instant",
    temperature=0
)

# 2. Set up the Retrieval Chain (The Bridge)
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectorstore.as_retriever()
)

# 3. Ask your question!
query = "Can you give me a 3-bullet point summary of this PDF?"
response = qa_chain.invoke(query)

print("--- ANSWER ---")
print(response["result"])
```
